```python
import pygame

pygame.init()

# Tamanho da tela:
window = pygame.display.set_mode([920, 600])
window_title = pygame.display.set_caption('Fute IGTI')

# Carregando imagem campo, jogador, bola:
field = pygame.image.load('assert/field.png')
goleiro1 = pygame.image.load('assert/goleiro1.png')

# Definindo a posição do 'goleiro1':
goleiro1_y = 300
goleiro1_moveup = False
goleiro1_movedown = False

goleiro2 = pygame.image.load('assert/goleiro2.png')
goleiro2_y = 300

ball = pygame.image.load('assert/ball.png')

# Posição da bola:
ball_x = 450
ball_y = 300
ball_dir = -1
ball_dir_y = 1

# Função para os movimentos do goleiro
def move_goleiro():
    global goleiro1_y

    if goleiro1_moveup:
        goleiro1_y -= 5
    else:
        goleiro1_y += 0

    if goleiro1_movedown:
        goleiro1_y += 5
    else:
        goleiro1_y += 0

    if goleiro1_y <= 0:
        goleiro1_y = 0
    elif goleiro1_y >= 575:
        goleiro1_y = 575

def move_goleiro2():
    global goleiro2_y
    goleiro2_y = ball_y

# Movendo a bola:
def move_ball():
    global ball_x
    global ball_y
    global ball_dir
    global ball_dir_y

    ball_x += ball_dir
    ball_y += ball_dir_y

    if ball_x < 120:
        if goleiro1_y < ball_y + 23:
            if goleiro1_y + 146 > ball_y:
                ball_dir *= -1

    if ball_x > 800:
        if goleiro2_y < ball_y + 23:
            if goleiro2_y + 146 > ball_y:
                ball_dir *= -1

    if ball_y > 685:
        ball_dir_y *= -1
    elif ball_y <= 0:
        ball_dir_y *= -1

# Passando os parametros:
def drow():
    # Posição do campo:
    window.blit(field, (0,0))
    window.blit(goleiro1, (50, goleiro1_y))
    window.blit(goleiro2, (820,goleiro2_y))
    window.blit(ball, (310,310))
    window.blit(ball, (ball_x, ball_y))


loop = True
while loop:
    for events in pygame.event.get():
        if events.type == pygame.QUIT:
            loop == False
        if events.type == pygame.KEYDOWN:
            if events.key == pygame.K_w:
                goleiro1_moveup = True
            if events.key == pygame.K_s:
                goleiro1_movedown = True
        if events.type == pygame.KEYUP:
            if events.key == pygame.K_w:
                goleiro1_moveup = False
            if events.key == pygame.K_s:
                goleiro1_movedown = False

    drow()
    move_ball()
    move_goleiro()
    move_goleiro2()
    pygame.display.update()

```
