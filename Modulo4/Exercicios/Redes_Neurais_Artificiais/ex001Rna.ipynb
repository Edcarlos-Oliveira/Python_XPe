```python
# Importando os modulos:
import pandas as pd
df = pd.read_csv('credit_data.csv')
```

```python
# Saída do 'df':
df.head()
```

```python
df = df.dropna()
```

```python
# Visualizando a quantidade de registros:
df.shape
```

```python
# Separação dos dados de entrada e saida:
x_credit = df[['income', 'age', 'loan']]
y_credit = df['default']
```

```python
# Normalizando os dados (organizando):
from sklearn.preprocessing import StandardScaler
standar_scaler = StandardScaler()
x_credit = standar_scaler.fit_transform(x_credit)
```

```python
# Saída da normalização dos dados:
x_credit
```

```python
# Separação dos dados de teste e treino
from sklearn.model_selection import train_test_split
x_credit_trein, x_credit_test, y_credit_trein, y_credit_test = train_test_split(x_credit, y_credit, test_size=0.25, random_state = 0)
```

```python
# Verificando como ficou a separação de 'x_credit_trein':
x_credit_trein.shape
```

```python
# Verificando como ficou a separação de 'y_credit_trein':
y_credit_trein.shape
```

```python
# Verificando como ficou a separação de 'x_credit_test e y_credit_test':
x_credit_test.shape, y_credit_test.shape
```

```python
# Aplicando o treinamento da RNA:
from sklearn.neural_network import MLPClassifier
rnc = MLPClassifier(max_iter = 1000, verbose = True, tol=0.0000100, hidden_layer_sizes=(20, 100)) # 'max_iter', máximo de interações, 'tol' da margem dos erros, 'hidden_layer_sizes', camadas ocultas 
rnc.fit(x_credit_trein, y_credit_trein)
```

```python
# Verificando as previsões dos dados treinados:
prev = rnc.predict(x_credit_test)

# Saída do 'prev':
prev
```

```python
# Verificando os dados treinados de 'y_credit_test':
y_credit_test
```

```python
# Verificando a acuracia do modelo treinado:
from sklearn.metrics import accuracy_score, classification_report

# Chamando a função:
accuracy_score(y_credit_test, prev)
```
